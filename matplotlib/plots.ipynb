```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories = np.array(["fruits", "vegetables", "dairy", "oats", "eggs", "sweets"])
values = np.array([2,4,1,6,2,1])

plt.title("daily Consumption")
plt.xlabel("food")
plt.ylabel("amount")

plt.bar(categories, values, color="red")
```

```python
plt.title("daily Consumption")
plt.xlabel("food")
plt.ylabel("amount")

plt.barh(categories, values, color="#2bd825")
# plt.plot(categories, values)
```

```python
colors = ["red", "lime", "#d7d7d7", "brown", "yellow", "blue"]
plt.pie(values, labels=categories, colors=colors, explode=[0,0,0.4,0,0,0.2], shadow=True)
plt.show()
```

```python
scores = np.random.normal(loc=75, scale=10, size=300)
scores = np.clip(scores, 0, 100)
```

```python
plt.hist(scores, edgecolor = 'black',
                 color= "lightgreen",
                 bins=12)

plt.title("scores distribution")
plt.xlabel("scores")
plt.ylabel("no. of students")
```
