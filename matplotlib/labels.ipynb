```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2021, 2022, 2023, 2024, 2025, 2026])
y1 = np.random.randint(1,100, 6)
y2 = np.random.randint(1,100, 6)
y3 = np.random.randint(1,100, 6)
```

```python
plt.title("Sales", fontsize="20",
                    fontweight='bold',
                    color="#0e04dc",
                    family='Arial')

plt.xlabel("Year",  fontsize="16",
                    fontweight='700',
                    color="#36e3f3",
                    family='Arial')

plt.ylabel("Revenue",  fontsize="16",
                    fontweight='700',
                    color="#36e3f3",
                    family='Arial')


plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```
