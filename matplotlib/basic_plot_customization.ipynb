```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
y = np.random.randint(1,30, 10)
y2 = np.random.randint(1,30, 10)
```

```python
# ms = markersize, mfc = markerfacecolor, mec = markeredgecolor
plt.plot(x,y, marker=".", ms='30', mfc='cyan', mec='lime')
plt.plot(x,y2, marker=".", ms='30', mfc='cyan', mec='lime')
plt.show()
```

```python
#line customizations
plt.plot(x,y, marker=".", ms='20', mfc='cyan', mec='blue',
              linestyle="dotted", #other options -> dashdot, solid, dotted, dashed
              linewidth=3,
              color='pink')
```
