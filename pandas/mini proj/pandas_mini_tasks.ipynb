```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "name" : ["samaira", "sanya", "sanyam", "abhinav", "soumil", "dhruv", "chakshu", "aditya","disha", "yashika"],
    "maths" : np.random.randint(50, 100, size=(10)),
    "science" : np.random.randint(50, 100, size=(10)),
    "english" : np.random.randint(50, 100, size=(10)),
    }, index=range(1,11))

df
```

```python
df['total marks'] = df[["maths", "science", "english"]].sum(axis = 1)
df
```

```python
df['percentage'] = np.round(df['total marks']/3, 2)
df
```

```python
maths_topper = df[df.maths > 80]
maths_topper
```

```python
fails = df[df[["maths", "science", "english"]].apply(lambda row: (row < 60).any(), axis=1)]
fails
```

```python
good_all =  df[df[["maths", "science", "english"]].apply(lambda row: (row > 70).all(), axis=1)]
good_all
```

```python
df_sorted_maths = df.sort_values("maths", ascending=False)
df_sorted_maths
```

```python
df_sorted_avg = df.sort_values("total marks", ascending=False)
df_sorted_avg
```
