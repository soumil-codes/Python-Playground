## 🟠 Task 5: IPL Dataset Analysis

```python
df = pd.read_csv("ipl/matches.csv")
df.head()
```

```python
df2 = pd.read_csv("ipl/deliveries.csv")
df2.loc[0:20, "total_runs"]
```

```python
df2.ball.value_counts()
df2.head(5)
```

```python
df2.batsman_runs.value_counts()
```

```python
top_run_scorrer = df2.groupby("batter").batsman_runs.sum()
top_run_scorrer = top_run_scorrer.sort_values(ascending=False)
```

```python
bowls_played = df2["batter"].value_counts()
bowls_played
```

```python
best_strike_rate = top_run_scorrer/bowls_played * 100
best_strike_rate.sort_values(ascending=False)
```

```python
sr = pd.DataFrame({
    "runs": top_run_scorrer,
    "balls" : bowls_played,
    "sr": best_strike_rate,
})

sr
```

```python
sr_filtered = sr[sr['balls'] > 200]
sr_filtered
```

```python
sr_filtered = sr_filtered.sort_values(by="sr", ascending=False)
sr_filtered.head(15)
```

```python
sr_filtered2 = sr_filtered[sr_filtered['runs'] > 3000]
sr_filtered2
```

```python
wickets_valid = df2[~df2['dismissal_kind'].isin(['run out', 'retired hurt'])]
wickets = wickets_valid.groupby('bowler').is_wicket.sum()
wickets
wickets.sort_values(ascending=False)
```

## 🟠 Task 6: Combine NumPy + Pandas

```python
data = np.random.randn(100, 3)
df = pd.DataFrame(data, columns=['a', 'b', 'c'])
df
```

```python
df.corr()
```

```python
z_index = (df - df.mean())/ df.std()
z_index
outliers = df[(z_index.abs() > 3).any(axis=1)]
outliers
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
outliers_iqr = df[
    ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)
]
outliers_iqr
```
