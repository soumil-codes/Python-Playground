## Messy Dataset Cleaning

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["samaira", "sanya", "sanyam", "abhinav", "soumil", 
             "dhruv", "chakshu", "aditya", "disha", "yashika"],
    
    "maths": [85, "90", 78, np.nan, "65", 85, "absent", 92, 88, 90],
    
    "science": [88, 76, "80", 70, np.nan, 88, 91, "85", 89, 76],
    
    "english": [90, 85, 88, "92", 75, np.nan, 84, 90, "eighty", 85]
})

# Add duplicate rows
df = pd.concat([df, df.iloc[[2, 5]]], ignore_index=True)

subjects = ['maths', 'science', 'english']
df
```

```python
df[df.isna().any(axis=1)]
```

```python
df.isna().sum(axis=1)
```

```python
df.fillna(0)
```

```python
df_clean = df[subjects].apply(pd.to_numeric, errors = "coerce")
df[subjects] = df_clean
df.fillna(0, inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.rename(columns={
    "maths" : "Mathematics",
    "science" : "Physics",
    "english" : "eng",
}, inplace=True)
df
```
