## Task 1: Student Marks Analyzer

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "name" : ["rahul", "sanya", "sanyam", "abhinav", "soumil", "dhruv", "chakshu", "aditya","disha", "yashika"],
    "maths" : np.random.randint(1, 100, size=(10)),
    "science" : np.random.randint(1, 100, size=(10)),
    "English" : np.random.randint(1, 100, size=(10)),
    }, index=range(1,11))

df

```

- Calculate average marks per student
- Find topper
- Find subject-wise average
- Add a “Pass/Fail” column

```python
avg_marks = np.round(df.iloc[:,1:4].sum(axis=1)/3, decimals=2)
avg_marks
```

```python
df["avg_marks"] = avg_marks
df
```

```python
topper = df.avg_marks.idxmax()
df.loc[topper]
```

```python
subject_wise_avg = df.loc[:, ['maths', 'science', 'English']].sum()/10
subject_wise_avg
```

```python
df["result"] = df[["maths", "science", "English"]].apply(lambda row : "pass" if (row > 40).all() else "fail", axis=1)
```
