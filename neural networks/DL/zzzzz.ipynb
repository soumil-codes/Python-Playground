```python
x = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,0])  # XOR truth table

def activation(z):
    return np.where(z >= 0, 1, 0)

# Initialize weights
np.random.seed(0)
W1 = np.random.randn(2, 2)   # (2 inputs в†’ 2 hidden neurons)
b1 = np.zeros((1, 2))
W2 = np.random.randn(2, 1)   # (2 hidden в†’ 1 output)
b2 = np.zeros((1, 1))

lr = 0.1
epochs = 100

for epoch in range(epochs):
    for i in range(len(x)):
        # Forward pass
        z1 = np.dot(x[i], W1) + b1
        a1 = activation(z1)
        z2 = np.dot(a1, W2) + b2
        a2 = activation(z2)

        # Error
        error = y[i] - a2

        # Update output layer
        W2 += lr * a1.reshape(-1,1) * error
        b2 += lr * error

        # Update hidden layer (perceptron style)
        W1 += lr * x[i].reshape(-1,1).dot(error * W2.T)
        b1 += lr * error * W2.T

# Test
for i in range(len(x)):
    z1 = np.dot(x[i], W1) + b1
    a1 = activation(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = activation(z2)
    print(f"{x[i]} -> {a2[0]}")
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# 1. Import MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize input data
x_train, x_test = x_train / 255.0, x_test / 255.0

# One-hot encode labels
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# 2. Build a feed forward neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')  # 10 classes for digits
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 3. Train FFNN for classification
model.fit(x_train, y_train_cat, epochs=30, batch_size=128)

# 4. Evaluate accuracy & performance metrics
y_pred_prob = model.predict(x_test)
y_pred = y_pred_prob.argmax(axis=1)

acc = accuracy_score(y_test, y_pred)
print("Test Accuracy:", acc)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
