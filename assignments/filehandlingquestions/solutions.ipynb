```python
def filereader(path):
    with open(path, 'r') as f:
        for line in f:
            print(line.strip())

filereader("sample.txt")
```

```python
def filewriter(path, arr):
    with open(path, 'w') as file:
        for string in arr:
            string += "\n"
            file.write(string)
        
arr = ["hello", "world", "example"]
path = "example.txt"

filewriter(path, arr)
filereader(path)

```

```python
def filecopy(source, destination):
    arr = []
    with open(source, 'r') as file:
        for line in file:
            arr.append(line.strip())
    
    with open(destination, 'w') as file:
        for line in arr:
            file.write(line + "\n")


filecopy("example.txt", "copy.txt")
filereader("copy.txt")
```
