```python
class Car:
    def __init__(self, model, make, year):
        self.model = model
        self.make = make
        self.year = year
        
    def info(self):
        print(f"model : {self.model} \nmake : {self.make} \nyear : {self.year}")
```

```python
car1 = Car("bmw", "gtr", 2011)
car1.info()
```
